```python
# Set Catalog and Schema context
spark.sql("USE CATALOG assignment;")
spark.sql("USE SCHEMA bronze;")

# Base path pointing to uploaded CSVs
base_path = "/Volumes/assignment/bronze/raw_data"

employee_src = f"{base_path}/Employee-Q1.csv"
department_src = f"{base_path}/Department-Q1.csv"
country_src    = f"{base_path}/Country-Q1.csv"

# Read datasets
employee_df = spark.read.option("header", True).csv(employee_src)
department_df = spark.read.option("header", True).csv(department_src)
country_df = spark.read.option("header", True).csv(country_src)

# Display to validate load
display(employee_df)
display(department_df)
display(country_df)
# Path to store bronze outputs
bronze_output_path = "/Volumes/assignment/bronze/processed"

# Write each dataset into bronze as raw CSV format
employee_df.write.mode("overwrite").option("header", True).csv(f"{bronze_output_path}/employee")
department_df.write.mode("overwrite").option("header", True).csv(f"{bronze_output_path}/department")
country_df.write.mode("overwrite").option("header", True).csv(f"{bronze_output_path}/country")

```

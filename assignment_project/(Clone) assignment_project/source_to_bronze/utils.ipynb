```python
import re
from pyspark.sql.functions import udf, current_date
from pyspark.sql.types import StringType


def camel_to_snake(name: str) -> str:
    s1 = re.sub(r'(.)([A-Z][a-z]+)', r'\1_\2', name)
    s2 = re.sub(r'([a-z0-9])([A-Z])', r'\1_\2', s1)
    return s2.lower()

camel_to_snake_udf = udf(camel_to_snake, StringType())

def convert_columns_to_snake_case(df):
    new_cols = [camel_to_snake(c) for c in df.columns]
    return df.toDF(*new_cols)


def add_load_date_column(df):
    return df.withColumn("load_date", current_date())

def add_at_load_date_column(df):
    return df.withColumn("at_load_date", current_date())
```
